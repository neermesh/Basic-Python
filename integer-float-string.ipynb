# 1.Numbers

#### _The interpreter acts as a simple calculator: you can type an expression at it and it will write the value. Expression syntax is straightforward: the operators +, -, * and / work just like in most other languages (for example, Pascal or C); parentheses (()) can be used for grouping. For example:_

```python
2 + 3
```

```python
35 - 4*7
```

```python
9 / 3  # division always returns a floating point number
```
